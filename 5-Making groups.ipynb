```python
import pandas as pd

# Load dataset file
ds = pd.read_csv("MEREGED_RAW_FINAL.csv")
ds.columns
```

```python
ds = ds.iloc[: , 1:]
ds.insert(0, 'Count_queries', -1)
ds.head(10)
```

```python
# df_cleaned4 = df_cleaned3.groupby(['groups']).mean()
ds2 = ds.groupby(['query']).agg({
                                                   'Input-U-Time': 'sum',
                                                   'Input-D-Time':'sum',

                                                   'Input-U-Size':'sum',
                                                   'Input-D-Size':'sum',

                                                   'Out-U-Time' : 'sum',
                                                   'Out-D-Time' : 'sum',

                                                   'Out-U-Size': 'sum',
                                                   'Out-D-Size': 'sum',

                                                   'Count_queries' : 'count',

                                             }).reset_index()
```

```python
ds2 = ds2.rename({
                                                   'Input-U-Time': 'sum Input-U-Time',
                                                   'Input-D-Time':'sum Input-D-Time',

                                                   'Input-U-Size':'sum Input-U-Size',
                                                   'Input-D-Size':'sum Input-D-Size',

                                                   'Out-U-Time' : 'sum Out-U-Time',
                                                   'Out-D-Time' : 'sum Out-D-Time',

                                                   'Out-U-Size': 'sum Out-U-Size',
                                                   'Out-D-Size': 'sum Out-D-Size',

                                                   'Count_queries' : 'count',

                    }, axis=1) 
```

```python
# saving the dataframe
ds2.to_csv('MEREGED_RAW_FINAL_GROUPED.csv')
```

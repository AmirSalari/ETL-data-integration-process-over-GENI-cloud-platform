```python
# ds_temp = [ds_1, ds_2, ds_3, ds_4, ds_5]
ds_temp = [ds_6]
  
ds_0 = pd.concat(ds_temp,ignore_index=True)

# ds.to_csv('grouped_merged.csv')

ds_0.info()
```

```python
# x_temp = ds.loc[:, ds.columns != "Label"]
# X = x_temp.loc[:, x_temp.columns != "groups"]

ds = ds_0.iloc[: , 2:]
ds.info()
```

```python
from sklearn.cluster import KMeans
clusters = 5

kmeans = KMeans(n_clusters = clusters, max_iter= 500, n_init=20)

cluster = kmeans.fit_predict(ds.iloc[: , 1:])
ds['Cluster'] = cluster
# ds.columns
```

```python
ds2 = ds[['query', 'count', 'Cluster']]
ds2.head()
```

```python
# applying groupby() function to
# group the data on team value.
gp = ds2.groupby('Cluster')
  
# Let's print the first entries
# in all the groups formed.
gp.first()
```

```python
for i in range(0, clusters):
  print("\nCluster: ", i, " is:\n", gp.get_group(i)['query'].value_counts())
# max()
```

```python
import pandas as pd

# Load dataset file
# xxx = pd.read_csv("final3.csv")
xxx = ds
domains_alphabet_unique = list(xxx['groups'].unique())
domains_alphabet = xxx['groups']

domains_number = []

for i in range(len(domains_alphabet)):
  if(domains_alphabet[i] in domains_alphabet_unique):
   key = domains_alphabet[i]
  #  domains_number.append(domains_alphabet_unique[key].index())
  # print(domains_alphabet_unique.index(key))
   domains_number.append(domains_alphabet_unique.index(key))

len(domains_number)
```
